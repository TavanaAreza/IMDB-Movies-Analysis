# Question 2 Part d
In this part we remove duplicated values and calculate profit and then analyze the most profitable movies.

```python
new_df_deleted_nan = new_df_deleted;
new_df_deleted_nan_dup = new_df_deleted_nan.drop_duplicates(keep = 'first')
new_df_deleted_nan_dup
```

```python
## Creating new columns as "profit"
new_df_deleted_nan_dup["profit"] = new_df_deleted_nan_dup.apply(lambda row:row.gross-row.budget,axis = 1)
new_df_deleted_nan_dup
```

```python
## Ploting profit against budget
plt.scatter(new_df_deleted_nan_dup["budget"],new_df_deleted_nan_dup["profit"])
plt.show
```

As we can see an outlier data and high leverage points cause that we can not interpret the normal data. Now we try to remove the outliers and then try to analyze the plot.


```python
## deleting outliers and high leverage points 
bounded_profit = new_df_deleted_nan_dup[new_df_deleted_nan_dup["profit"] 
                                        > -3 * np.power(10,8) ]["profit"]

bounded_budget = new_df_deleted_nan_dup[new_df_deleted_nan_dup["budget"] 
                                        < 3.5 * np.power(10,8)]["budget"]

plt.scatter(bounded_budget,bounded_profit)
plt.show
```

With new bounding in dataset and remove outliers and high leverage points, the plot is now more interpretable.\
We can conclude that there is a weak trend that more budget that spent for a film, the making profit for that will become more harder in most cases.\
Also there is many films that has negative profit and had disadvantage. It shows that more founding and investigating on a film doesn't cause more profit and gain.\
Actually there are some films that with high founding and budget that had high proftit too.\
Also most cases were profitable in this dataset and the profit is higher than 0.
And also there are some films that with low founding and budget, had high profit that it shows they were successful in sale and atracted people more.\

# Question 2 Part g
In this part we investigate whether folks rating has effect on gross or not. I do it plotting and regression analysis to figure this out.

```python
## first we plot the scatter plot to check whether there is a relation between
## gross and people rating or not
## we use "new_df_deleted_nan_dup["budget"]" variable in this part which Nan 
## and duplicated values are removed
plt.scatter(new_df_deleted_nan_dup["imdb_score"],new_df_deleted_nan_dup["gross"])
```

As we can see in above, it seems there is a weak relation between imdb_score and gross. Now in the next part we try to fit a line to the data to check this out more.

```python
## fitting a regression line to the above plot
sns.lmplot(x='imdb_score',y='gross',data = new_df_deleted_nan_dup, fit_reg=True,
           line_kws={'color': 'red'}) 
```

As we can see in regression line in above, we can conclude there is a weak relation between imdb score and gross. For the next part I do a hypothesis testing and calculate the Chi-Squard testing to figure out the conclusion more carefully.

```python
## doing Chi-Squared test to see whether folks rating has effect on gross or not
table = [new_df_deleted_nan_dup["gross"],new_df_deleted_nan_dup["imdb_score"]]
stat, p, dof, expected = chi2_contingency(table)
print('stat=%f, p=%f' % (stat, p))
if p > 0.05:
	print('"gross" and "imdb_score" parameters are independent')
else:
	print('"gross" and "imdb_score" parameters are dependent')
```

As we can see in above t test, the stat is too large and the p-value is too small so as we concluded, these two parameters are dependent. We test to prove it statistically significant.
